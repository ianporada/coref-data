```python
from pathlib import Path

import xml.etree.ElementTree as ET

import conll_transform
from datasets import Dataset, DatasetDict
```

```python
fname = Path("/Users/ianporada/Research/data/push_to_hub/phrase_detectives/PD3.0_Train_dev/gold/gnome/CONLL/getty_Cabinet_mas.CONLL")
```

```python
def remove_nonreferring(sentences):
    for s in sentences:
        for row in s:
            coref_bounds = row[-1].split("|")
            # remove bounds with N (non-referring)
            coref_bounds = filter(lambda x: 'N' not in x, coref_bounds)
            coref_bounds = map(lambda x: x.replace("R", ""), coref_bounds)
            row[-1] = "|".join(coref_bounds)
    return sentences


def read_conll_file(fname):
    docs = []
    for doc_name, sentences in conll_transform.read_file(fname).items():
        sentences = remove_nonreferring(sentences)
        try:
            coref_chains = conll_transform.compute_chains(sentences)
        except:
            print(f"{doc_name} compute chains failed at file {fname}")
            raise
        docs.append({
            "doc_name": doc_name,
            "sentences": sentences,
            "coref_chains": coref_chains,
        })
    return docs
```
